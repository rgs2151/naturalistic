```python
from ultralytics import YOLO
```

```python
# Load a COCO-pretrained YOLO11n model
model = YOLO("yolo11m.pt")
```

```python
results = model.track("bin/bangbangyouredead_v4-1.mp4", show=False, save=True)
```

```python
data = {
    "class": [],
    "class_conf": [],
    "bbox_xywh": [],
}

for r in results:
    data["class"].append(r.boxes.cls.tolist())
    data["class_conf"].append(r.boxes.conf.tolist())
    data["bbox_xywh"].append(r.boxes.xywh.tolist())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)

df.to_parquet("bin/bangbangyouredead_v4-1.parquet")
```
